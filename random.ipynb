# Understanding Probability
Probability of Events
The probability of an event happening is a number between 0 and 1, where 0 means the event is impossible, and 1 means the event is certain. For example:

The probability of rolling a 3 on a fair die is 1/6, because there are 6 equally likely outcomes.

The probability of flipping heads on a fair coin is 0.5, because the two possible outcomes are equally likely.

Simulating Multiple Coin Flips
Let's simulate flipping a coin 100 times to observe the probability of getting heads and tails. We'll track how the proportion of heads changes as we flip the coin more times.

```python
# Simulate multiple coin flips
num_flips = 100
results = [random.choice(['Heads', 'Tails']) for _ in range(num_flips)]

# Calculate proportions
heads_count = results.count('Heads')
tails_count = results.count('Tails')

print(f"Number of Heads: {heads_count}")
print(f"Number of Tails: {tails_count}")
print(f"Proportion of Heads: {heads_count/num_flips}")
print(f"Proportion of Tails: {tails_count/num_flips}")
```

If we roll a die multiple times, we can compute the frequency of each result. This gives us an insight into the distribution of outcomes.

```python
# Simulate multiple dice rolls
num_rolls = 1000
rolls = [random.randint(1, 6) for _ in range(num_rolls)]

# Count the frequency of each outcome
frequencies = [rolls.count(i) for i in range(1, 7)]

# Display results
for i, freq in enumerate(frequencies, 1):
    print(f"Side {i}: {freq} times, Proportion: {freq / num_rolls}")
```

This simulates 1000 dice rolls and shows how often each side of the die appears. Since the die is fair, we expect the proportions to be roughly equal, but due to randomness, some variations will occur.

# Understanding Random Variables and Distributions
In statistics, random variables are fundamental in describing random phenomena. We can model random variables using different probability distributions.

Uniform Distribution
The uniform distribution is the simplest distribution, where all outcomes are equally likely.

```python
uniform_data = np.random.uniform(0, 1, 1000)
sns.histplot(uniform_data, kde=True, bins=20)
plt.title("Uniform Distribution")
plt.show()
```

 Normal Distribution
A normal (Gaussian) distribution is one of the most widely used distributions, characterized by its bell shape.

```python
normal_data = np.random.normal(0, 1, 1000)
sns.histplot(normal_data, kde=True, bins=30)
plt.title("Normal Disturbution")
plt.show()
```

Exponential Distribution
Exponential distribution models the time between events in a Poisson process, often used in survival analysis.

```python
# Generate random data from an exponential distribution with lambda=1
exponential_data = np.random.exponential(1, 1000)

# Plot the histogram
sns.histplot(exponential_data, kde=True, bins=30)
plt.title("Exponential Distribution")
plt.show()
```

Binomial Distribution
The Binomial Distribution models the number of successes in a fixed number of independent trials of a binary experiment (success or failure). It's widely used when you want to model situations where there are two possible outcomes (e.g., heads or tails, success or failure).

```python
# Generate random data from a binomial distribution
n, p = 10, 0.5  # 10 trials, 50% chance of success
binomial_data = np.random.binomial(n, p, 1000)

# Plot the histogram with a KDE
sns.histplot(binomial_data, kde=True, bins=20)
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.show()
```

Poisson Distribution
The Poisson Distribution models the number of events occurring within a fixed interval of time or space, given that the events happen with a constant average rate and independently of one another. It's often used in queueing theory, traffic modeling, and incident counting.

```python
# Generate random data from a Poisson distribution
lambda_param = 3  # Average number of events per interval
poisson_data = np.random.poisson(lambda_param, 1000)

# Plot the histogram with a KDE
sns.histplot(poisson_data, kde=True, bins=20)
plt.title("Poisson Distribution (λ=3)")
plt.xlabel("Number of Events")
plt.ylabel("Frequency")
plt.show()
```

Gamma Distribution
The Gamma Distribution is a two-parameter family of continuous probability distributions. It is used to model the waiting time until a Poisson process reaches a certain number of events.

```python
# Generate random data from a Gamma distribution
k, theta = 2, 2  # shape = 2, scale = 2
gamma_data = np.random.gamma(k, theta, 1000)

# Plot the histogram with a KDE
sns.histplot(gamma_data, kde=True, bins=30)
plt.title("Gamma Distribution (k=2, θ=2)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

# Statistical Sampling and Random Sampling Techniques

Sampling is a crucial concept in statistics and data science. It's often impossible or impractical to collect data from an entire population, so sampling allows us to estimate characteristics of a population based on a smaller subset. There are three Types of Sampling Techniques

Simple Random Sampling (SRS)

Sampling with Replacement

Stratified Sampling

In Simple Random Sampling (SRS), every element in the population has an equal chance of being selected. This is the most basic type of sampling and serves as the foundation for many other sampling techniques.

```python
population = list(range(1, 101))  # A population of numbers from 1 to 100
sample = random.sample(population, 10)  # Sample 10 random elements
print(f"Simple Random Sample: {sample}")
```

Sampling with Replacement
Sampling with replacement means that after an element is selected, it is placed back into the population, allowing it to be selected again.

```python
sample_with_replacement = random.choices(population, k=10)  # Sample 10 with replacement
print(f"Sample with Replacement: {sample_with_replacement}")
```

Stratified Sampling
Stratified sampling involves dividing the population into strata (groups) and sampling from each group.

```python
# Create a population with different strata
strata = {
    'Group A': [1, 2, 3, 4, 5],
    'Group B': [6, 7, 8, 9, 10],
}

# Take a sample from each group
sample_stratified = {group: random.sample(members, 2) for group, members in strata.items()}
print(f"Stratified Sampling: {sample_stratified}")
```

# The law of large numbers 

The Law of Large Numbers (LLN) is a fundamental concept in probability and statistics. It states that as the number of trials in an experiment increases, the average of the observed outcomes will converge to the expected value (or theoretical mean) of the random variable.

In simpler terms:

If you repeat a random experiment (like flipping a coin or rolling a die) many times, the average of your results will get closer and closer to the expected probability of each outcome.

For example, when flipping a fair coin, the expected probability of getting heads is 0.5 (50%). If you flip the coin a few times, the proportion of heads might be quite different from 0.5 due to random variation. However, as you flip the coin more and more times, the proportion of heads should get closer to 0.5.

```python
# Simulate a large number of coin flips
num_flips = 10000
results = [random.choice(['Heads', 'Tails']) for _ in range(num_flips)]

# Calculate the running proportion of heads
running_heads = [results[:i].count('Heads') / i for i in range(1, num_flips + 1)]

# Plot the running proportion
plt.plot(running_heads)
plt.axhline(0.5, color='red', linestyle='--', label="Expected Probability (0.5)")
plt.title("Running Proportion of Heads in Coin Flips")
plt.xlabel("Number of Flips")
plt.ylabel("Proportion of Heads")
plt.legend()
plt.show()
```

This code simulates flipping a coin 10,000 times and tracks the running proportion of heads. As we increase the number of flips, the proportion of heads should converge toward 0.5, demonstrating the Law of Large Numbers.

You can extend the idea of the Law of Large Numbers to other random events as well. For example:

Dice Rolls:

If you roll a fair six-sided die many times, the proportion of each face (1 through 6) will converge toward 1/6. Each number has a 1/6 probability, so after many rolls, the proportion of rolls that result in each number should be approximately 1/6.

```python
# Simulate dice rolls
num_rolls = 10000
dice_results = [random.randint(1, 6) for _ in range(num_rolls)]

# Calculate the running proportions of each face
running_proportions = {i: [dice_results[:j].count(i) / j for j in range(1, num_rolls + 1)] for i in range(1, 7)}

# Plot the running proportions
for i in range(1, 7):
    plt.plot(running_proportions[i], label=f"Face {i}")
    
plt.axhline(1/6, color='black', linestyle='--', label="Expected Proportion (1/6)")
plt.title("Running Proportions of Dice Rolls")
plt.xlabel("Number of Rolls")
plt.ylabel("Proportion of Each Face")
plt.legend()
plt.show()
```

# Monte Carlo simulations 

These are a powerful technique used to solve problems that might be deterministic in theory but are too complex or intractable to solve exactly. These simulations rely on the use of randomness (random sampling) to approximate a solution to a problem.

What is Monte Carlo Simulation?
A Monte Carlo simulation involves:

Repeated random sampling of input parameters.

Using these random samples to simulate a process.

Aggregating the results to estimate the output of the process.

This approach is useful in scenarios where analytical solutions are hard to find, such as in financial modeling, risk analysis, and scientific computing.

Key Steps in a Monte Carlo Simulation:
Define a probabilistic model for the random variables (e.g., dice rolls, stock prices).

Simulate a large number of possible outcomes by repeatedly sampling from the model.

Aggregate the results to estimate a value (e.g., an average or probability).




monte carlo to calculate PI

```python
def estimate_pi(num_points=100000):
    inside_circle = 0 
    x_inside, y_inside = [], [] 
    x_outside, y_outside = [], [] 
    
    for _ in range(num_points):
        x, y = random.random(), random.random() 
        if x**2 + y**2 <= 1:
            inside_circle += 1 
            x_inside.append(x)
            y_inside.append(y)
        else:
            x_outside.append(x)
            y_outside.append(y)
            pi_estimate = (inside_circle / num_points) * 4
    return pi_estimate, x_inside, y_inside, x_outside, y_outside

# Estimate Pi and get the points for visualization
pi_estimate, x_inside, y_inside, x_outside, y_outside = estimate_pi(10000)

# Print the estimate of Pi
print(f"Estimated Pi: {pi_estimate}")

# Plot the results
plt.figure(figsize=(6,6))
plt.scatter(x_inside, y_inside, color='blue', s=1, label="Points inside Circle")
plt.scatter(x_outside, y_outside, color='red', s=1, label="Points outside Circle")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Monte Carlo Simulation of Pi Estimation")
plt.legend()
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

A random walk is another example of a process that can be simulated using Monte Carlo methods. In a random walk, you start at an initial position and take steps in random directions (e.g., left or right). This model is often used to represent stock prices, particle movement, and other random processes.

```python
def random_walk(steps=1000):
    position = 0
    positions = [position]

    # Simulate the random walk
    for _ in range(steps):
        step = random.choice([-1, 1])  # Random step, either left (-1) or right (+1)
        position += step
        positions.append(position)

    return positions

# Simulate a random walk
walk = random_walk(1000)

# Plot the random walk
plt.plot(walk)
plt.title("1D Random Walk Simulation")
plt.xlabel("Steps")
plt.ylabel("Position")
plt.show()
```

Monte Carlo simulations can be used for game simulations. In this case, we will simulate a dice roll game between two players, where each player rolls a die, and the player with the higher roll wins.

```python
def dice_game(num_simulations = 10000):
    player1_wins = 0 
    player2_wins = 0 
    ties = 0 
    for _ in range(num_simulations):
        player1_roll = random.randint(1, 6)  # Simulate dice roll for Player 1
        player2_roll = random.randint(1, 6)  # Simulate dice roll for Player 2

        if player1_roll > player2_roll:
            player1_wins += 1
        elif player2_roll > player1_roll:
            player2_wins += 1
        else:
            ties += 1

    return player1_wins, player2_wins, ties

# Run the dice game simulation
player1_wins, player2_wins, ties = dice_game(10000)

# Display the results
print(f"Player 1 Wins: {player1_wins} times")
print(f"Player 2 Wins: {player2_wins} times")
print(f"Ties: {ties} times")
```
