# Random Events: Coin Flip and Dice Roll
One of the most basic forms of randomness is flipping a coin or rolling a die. These simple events are easy to simulate and can provide a good introduction to probability.

```python
coin_flip = random.choice(['heads', 'tails'])
print(F"Coin Flip: {coin_flip}")
```

A standard six-sided die has six possible outcomes, each with an equal probability of 1/6. 

```python
die_roll = random.randint(1, 6)
print(f"Die Roll: {die_roll}")
```

# Understanding Probability
Probability of Events
The probability of an event happening is a number between 0 and 1, where 0 means the event is impossible, and 1 means the event is certain. For example:

The probability of rolling a 3 on a fair die is 1/6, because there are 6 equally likely outcomes.

The probability of flipping heads on a fair coin is 0.5, because the two possible outcomes are equally likely.

Simulating Multiple Coin Flips
Let's simulate flipping a coin 100 times to observe the probability of getting heads and tails. We'll track how the proportion of heads changes as we flip the coin more times.

```python
# Simulate multiple coin flips
num_flips = 100
results = [random.choice(['Heads', 'Tails']) for _ in range(num_flips)]

# Calculate proportions
heads_count = results.count('Heads')
tails_count = results.count('Tails')

print(f"Number of Heads: {heads_count}")
print(f"Number of Tails: {tails_count}")
print(f"Proportion of Heads: {heads_count/num_flips}")
print(f"Proportion of Tails: {tails_count/num_flips}")
```

If we roll a die multiple times, we can compute the frequency of each result. This gives us an insight into the distribution of outcomes.

```python
# Simulate multiple dice rolls
num_rolls = 1000
rolls = [random.randint(1, 6) for _ in range(num_rolls)]

# Count the frequency of each outcome
frequencies = [rolls.count(i) for i in range(1, 7)]

# Display results
for i, freq in enumerate(frequencies, 1):
    print(f"Side {i}: {freq} times, Proportion: {freq / num_rolls}")
```

This simulates 1000 dice rolls and shows how often each side of the die appears. Since the die is fair, we expect the proportions to be roughly equal, but due to randomness, some variations will occur.

# Understanding Random Variables and Distributions
In statistics, random variables are fundamental in describing random phenomena. We can model random variables using different probability distributions.

Uniform Distribution
The uniform distribution is the simplest distribution, where all outcomes are equally likely.

```python
uniform_data = np.random.uniform(0, 1, 1000)
sns.histplot(uniform_data, kde=True, bins=20)
plt.title("Uniform Distribution")
plt.show()
```

 Normal Distribution
A normal (Gaussian) distribution is one of the most widely used distributions, characterized by its bell shape.

```python
normal_data = np.random.normal(0, 1, 1000)
sns.histplot(normal_data, kde=True, bins=30)
plt.title("Normal Disturbution")
plt.show()
```

Exponential Distribution
Exponential distribution models the time between events in a Poisson process, often used in survival analysis.

```python
# Generate random data from an exponential distribution with lambda=1
exponential_data = np.random.exponential(1, 1000)

# Plot the histogram
sns.histplot(exponential_data, kde=True, bins=30)
plt.title("Exponential Distribution")
plt.show()
```

# Statistical Sampling and Random Sampling Techniques
Random sampling is an essential technique in data science and statistics for selecting representative samples from a population.

Simple Random Sampling
In simple random sampling, each element in the population has an equal chance of being selected.

```python
population = list(range(1, 101))  # A population of numbers from 1 to 100
sample = random.sample(population, 10)  # Sample 10 random elements
print(f"Simple Random Sample: {sample}")
```

Sampling with Replacement
Sampling with replacement means that after an element is selected, it is placed back into the population, allowing it to be selected again.

```python
sample_with_replacement = random.choices(population, k=10)  # Sample 10 with replacement
print(f"Sample with Replacement: {sample_with_replacement}")
```

Stratified Sampling
Stratified sampling involves dividing the population into strata (groups) and sampling from each group.

```python
# Create a population with different strata
strata = {
    'Group A': [1, 2, 3, 4, 5],
    'Group B': [6, 7, 8, 9, 10],
}

# Take a sample from each group
sample_stratified = {group: random.sample(members, 2) for group, members in strata.items()}
print(f"Stratified Sampling: {sample_stratified}")
```
